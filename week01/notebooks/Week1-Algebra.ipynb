### <u> ðŸŸ¨ Difference of Squares Identity </u>

The **difference of squares** states:

$$
a^2 - b^2 = (a - b)(a + b)
$$

---

### ðŸ”Ž Why It Works

Expand the right-hand side:

$$
(a - b)(a + b) = a^2 + ab - ab - b^2 = a^2 - b^2
$$

So:
$$
a^2 - b^2 = (a - b)(a + b)
$$

---

### ðŸ§  When to Use It

You can apply this identity to factor expressions like:

$
x^2 - 9 = (x - 3)(x + 3)
$

$
4x^2 - 25 = (2x - 5)(2x + 5)
$

$
9a^2 - 1 = (3a - 1)(3a + 1)
$

> **Note**: This does **not** work for sums of squares like $a^2 + b^2$.

---

### âœ… Summary

- Only applies to **differences** of perfect squares  
- Useful for factoring, solving equations, simplifying expressions  
- A special case of the product of conjugates




### <u> ðŸ“˜ The Quadratic Formula </u>

The **quadratic formula** is used to solve any equation of the form:

$$
ax^2 + bx + c = 0
$$

Where:
- $a$, $b$, and $c$ are constants (with $a \ne 0$)
- $x$ is the unknown variable

---

### ðŸ§® The Formula

$$
x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}
$$

This formula gives you the **solutions (roots)** of the quadratic equation.

---

### ðŸ” Understanding Each Part

| Part                          | Meaning                                               |
|-------------------------------|-------------------------------------------------------|
| $$ -b $$                      | Opposite of the linear coefficient                    |
| $\sqrt{b^2 - 4ac}$            | The **discriminant**                                  |
| $$ 2a $$                      | Denominator from solving by completing the square     |
| $$ \pm $$                     | Indicates **two solutions** (plus and minus)          |

---

### ðŸ“ The Discriminant

The discriminant is the expression inside the square root:

$$
D = b^2 - 4ac
$$

It tells you the **type of roots** the equation has:

- If $D > 0$: Two **real and distinct** solutions  
- If $D = 0$: One **real repeated** (or double) root  
- If $D < 0$: Two **complex (imaginary)** solutions  

---

### âœ… Example

Solve the equation:

$$
x^2 - 5x + 6 = 0
$$

Here:
- $a = 1$
- $b = -5$
- $c = 6$

Plug into the quadratic formula:

$$
x = \frac{-(-5) \pm \sqrt{(-5)^2 - 4(1)(6)}}{2(1)} = \frac{5 \pm \sqrt{25 - 24}}{2} = \frac{5 \pm \sqrt{1}}{2}
$$

Now compute both solutions:

$
x = \frac{5 + 1}{2} = 3
$

$
x = \frac{5 - 1}{2} = 2
$

---

### ðŸ§¾ Final Answer

$
x = 2 \quad \text{or} \quad x = 3
$

These are the two real roots of the equation.

```python
#Exercise Set 5: Using SymPy for Algebra

#Define symbolic variable
x = sp.Symbol('x')

#Solve equations symbolically
eq1 = sp.Eq(3*x + 7, 22)
solution1 = sp.solve(eq1, x)
print(f"3x + 7 = 22: x = {solution1}")

eq2 = sp.Eq(2*(x - 3), 4*x + 8)
solution2 = sp.solve(eq2, x)
print(f"2(x - 3) = 4x + 8: x = {solution2}")

eq3 = x**2 - 5*x + 6
solution3 = sp.solve(eq3, x)
print(f"xÂ² - 5x + 6 = 0: x = {solution3}")

#Expand and factor expressions
expr = (x + 3)**2
expanded = sp.expand(expr)
print(f"(x + 3)Â² = {expanded}")

expr2 = x**2 + 6*x + 9
factored = sp.factor(expr2)
print(f"xÂ² + 6x + 9 = {factored}")
```

```python
#Exercise Set 6: Functions and Their Properties

#Define and explore functions
def f(x):
    return 2*x + 3

def g(x):
    return x**2 - 1

def h(x):
    return math.sin(x)

#Evaluate functions at different points
x_values = [-2, -1, 0, 1, 2]
print("x\tf(x)\tg(x)\th(x)")
print("-" * 25)
for x in x_values:
    print(f"{x}\t{f(x)}\t{g(x)}\t{h(x):.3f}")

#Plot multiple functions
x_plot = np.linspace(-3, 3, 100)
plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.plot(x_plot, 2*x_plot + 3, 'r-')
plt.title('f(x) = 2x + 3')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(x_plot, x_plot**2 - 1, 'g-')
plt.title('g(x) = xÂ² - 1')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(x_plot, np.sin(x_plot), 'b-')
plt.title('h(x) = sin(x)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
