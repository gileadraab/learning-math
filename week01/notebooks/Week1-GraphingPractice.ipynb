```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Defining the Functions
functions = [
    ("Linear", lambda x: 2*x + 1),
    ("Quadratic", lambda x: x**2 - 3),
    ("Cubic", lambda x: x**3 - x),
    ("Square Root", lambda x: np.sqrt(np.abs(x))),
    ("Exponential", lambda x: np.exp(x/2)),
    ("Logarithmic", lambda x: np.log(np.abs(x) + 1))
]
# Creates a list of 6 tuples, each containing a name for the plot and a lambda function.
# NumPy functions ensure vectorized, elementwise computation.
# np.abs ensures valid input for sqrt and log.

# 2. Creating the X-values
x = np.linspace(-3, 3, 100)
# Generates 100 evenly spaced values between -3 and 3.
# Produces a 1D NumPy array for plotting.

# 3. Creating the Figure and Axes
overlay_fig, axes = plt.subplots(2, 3, figsize=(15, 10), layout="constrained")
# Creates a 2×3 grid of subplots (6 total plots) inside a figure sized 15×10 inches.
# fig is the container, axes is a 2D NumPy array of Axes objects.

# 4. Flattening Axes for Easy Looping
axes = axes.flatten()
# Converts axes from a (2,3) array into a flat array for simple iteration.

# 5. Plotting Each Function
for i, (name, func) in enumerate(functions):
    axes[i].plot(x, func(x))
    axes[i].set_title(name)
    axes[i].grid(True)
    axes[i].set_xlabel('x')
    axes[i].set_ylabel('y')
# Iterates through each function, calculates y-values, and plots them.
# Adds title, grid, and axis labels to each subplot.


plt.show()
# Displays the final figure.
```
