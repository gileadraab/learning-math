```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
from sympy import symbols, solve, diff, integrate
import math
import sys
```

```python
# Basic arithmetic in Python
print("=== Basic Arithmetic ===")
a = 15
b = 4

print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")  # Floor division
print(f"{a} % {b} = {a % b}")    # Remainder
print(f"{a} ** {b} = {a ** b}")  # Exponentiation
```

```python
# Practice with math functions
print("=== Math Module Functions ===")
x = 16
y = -7.5

print(f"Square root of {x}: {math.sqrt(x)}")
print(f"Absolute value of {y}: {abs(y)}")
print(f"Ceiling of {y}: {math.ceil(y)}")
print(f"Floor of {y}: {math.floor(y)}")
print(f"π (pi): {math.pi}")
print(f"e: {math.e}")
print(f"sin(π/2): {math.sin(math.pi/2)}")
print(f"log(10): {math.log10(100)}")
```

```python
# Creating your first mathematical plot
x = np.linspace(-2, 2, 100) # 100 points from -2 to 2
y = x**2

plt.figure(figsize=(8, 6)) # Create a new figure with a specified size (8 inches wide by 6 inches tall)
plt.plot(x, y, 'b-', label='y = x²') # Plot the x and y values as a blue line ('b-'), with a label for the legend
plt.grid(True)  # Show a grid in the background to help read the values
plt.xlabel('x') # Label the x-axis
plt.ylabel('y') # Label the y-axis
plt.title('My First Mathematical Plot') # Set the title of the plot
plt.legend()# Display the legend (uses the label from plt.plot)

plt.show()# Display the plot window
```

```python
#Exercise: Try plotting y = x³, y = sin(x), y = e^x**

x = np.linspace(-2, 2, 200)
plt.figure(figsize=(6, 12)) 

#y = x^3
plt.subplot(3, 1, 1)  # (nrows, ncols, index)
plt.plot(x, x**3, label='y = x³')
plt.title('y = x³')
plt.grid(True)
plt.legend()

# y = sin(x)
plt.subplot(3, 1, 2)
plt.plot(x, np.sin(x), label='y = sin(x)', color='g')
plt.title('y = sin(x)')
plt.grid(True)
plt.legend()

# y = e^x
plt.subplot(3, 1, 3)
plt.plot(x, np.exp(x), label='y = e^x', color='r')
plt.title('y = e^x')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
